```python
import matplotlib.pyplot as plt
import numpy as np

# plt.style.use('_mpl-gallery-nogrid')

# make data with uneven sampling in x
x = np.loadtxt('data_hi.dat',usecols=(0),unpack=True)
X, Y = np.meshgrid(x, np.loadtxt('data_hi.dat',usecols=(1),unpack=True))
Z = np.loadtxt('data_hi.dat',usecols=(2),unpack=True)

# plot
fig, ax = plt.subplots()

ax.pcolormesh(X, Y, Z, vmin=-0.5, vmax=1.0)

plt.show()
```
